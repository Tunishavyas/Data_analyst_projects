```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("students.csv")
print(df.head())
```

# Drop unnamed column


```python
df=df.drop("Unnamed: 0",axis=1)              #1 for vertical axis
```

```python
print(df.head())
```

# change weekly stdy hours

```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("05-oct","5-10")
df.head()
```

# Gender distribution

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])     #displaying exact value of males and females
plt.title("Gender Distribution")
plt.show()
```

# from the above chart we have analysed that:
the number of females in the data is more than the no of males

```python
gp=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gp)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gp,annot=True)             #annot is for showing the values
plt.title("Relationship Between Parent's Education and Student Score")
plt.show()
```

# from the above chart we conclude that the 
education of the parents have a good impact on their scores

```python
gp1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gp1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gp1,annot=True)             #annot is for showing the values
plt.title("Relationship Between Parent's Marital Status and Student Score")
plt.show()
```

# from the above chart we conclude that 
there is no/neglible impact on the student's score due to their parents marital status

# Distribution of Ethnic Groups

```python
groupA=df.loc[(df['EthnicGroup']== "group A")].count()                              
groupB=df.loc[(df['EthnicGroup']== "group B")].count()
groupC=df.loc[(df['EthnicGroup']== "group C")].count()
groupD=df.loc[(df['EthnicGroup']== "group D")].count()
groupE=df.loc[(df['EthnicGroup']== "group E")].count()
#print(groupA["EthnicGroup"])

l=["group A","group B","group C","group D","group E"]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
